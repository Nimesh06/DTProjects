# Project 1. Winequality-Red. data set

```python
# Importing required librarys
import pandas as pd

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# creating Dataframe
df = pd.read_csv('winequality_red.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# finding out the missing values
df.isna().sum()
# there are no missing values inthe data set.
```

```python
# The distribution of the quality variable. 
df.hist('quality')
```

```python
# Finding out the correlations between the variables,to find out the relationships between variables.
# To find out the variables that are strongly correlated to quality.
corr = df.corr()
plt.subplots(figsize=(15,10))
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

```python
# To compare the effectiveness of different classification techniques, the output variable is changed to a binary output.
# Defined a bottle of wine as good quality if it had a quality score of 7 or higher, and if it had a score of less than 7, it is defined as bad quality.
```

```python
df['goodquality'] = [1 if x >= 7 else 0 for x in df['quality']]
```

```python
df.head()
```

```python
# After converting the output variable into binary output, the feature variables (x) and the target variable (y) are seperated into different dataframes.
```

```python
x = df.drop(['quality','goodquality'], axis = 1)
y = df['goodquality']
```

```python
# To see the proportion of good quality and bad quality wines(basically the number).
# "1" represents the good quality and "0" represents bad quality.
df['goodquality'].value_counts()
```

```python
# Standardizing Feature Variables
# preparinr the data for modelling.
# The first thing is to standardize the data.(transforming the data so that its distribution will have a mean of 0 and a standard deviation of 1)(to equalize the range of the data)
```

```python
# Normalizing the variables
from sklearn.preprocessing import StandardScaler
x_features = x
x = StandardScaler().fit_transform(x)
```

```python
# Spliting the data into a training and test set to cross-validate the models and determining their effectiveness.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.25, random_state=0)
```

```python
# Modelling
```

```python
# Model 1 ----------Decision Tree Classifier----------
```

```python
# importing the reqired librarys
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
# Applying the model
model1 = DecisionTreeClassifier(random_state=1)
model1.fit(x_train, y_train)
y_pred1 = model1.predict(x_test)
print(classification_report(y_test, y_pred1))
print("accuracy_score:",accuracy_score(y_test,y_pred1))
```

```python
# Model 2 ----------Random Forest-----------
```

```python
# Importing required librarys
from sklearn.ensemble import RandomForestClassifier
```

```python
# Applying the model
model2 = RandomForestClassifier(random_state=1)
model2.fit(x_train, y_train)
y_pred2 = model2.predict(x_test)
print(classification_report(y_test, y_pred2))
print("accuracy_score:",accuracy_score(y_test,y_pred2))
```

```python
# Model 3 ----------Logistic Regression---------
```

```python
# Importing required librarys
from sklearn.linear_model import LogisticRegression
```

```python
model3 = LogisticRegression()
model3.fit(x_train,y_train)
y_pred3 = model3.predict(x_test)
print(classification_report(y_test, y_pred3))
print("accuracy_score:",accuracy_score(y_test,y_pred3))
```

```python
# Feature Importance
```

```python
# 1. Decision Tree Model
```

```python
feat_importances = pd.Series(model1.feature_importances_, index=x_features.columns)
feat_importances.nlargest(25).plot(kind='barh',figsize=(10,10))
```

```python
# 2. Random Forest Model
```

```python
feat_importances = pd.Series(model2.feature_importances_, index=x_features.columns)
feat_importances.nlargest(25).plot(kind='barh',figsize=(10,10))
```

The above graphed feature importance based on the Random Forest model and the DEcision Tree model. While they slightly vary, the top 2 features are the same: alcohol, volatile acidity, and 1 feature sulphates has resemblance also.

```python
# AUC-ROC Curve
```

```python
# importing required librarys
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
# AUC-ROC for-------Random Forest-------
y_pred_prob = model2.predict_proba(x_test)[:,1]
y_pred_prob
```

```python
fpr,tpr,thresholds=roc_curve(y_test,y_pred_prob)
print('FPR values:',fpr)
print('TPR values:',tpr)
print('thresholds values:',thresholds)
```

```python
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='Random Forest')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Random Forest')
plt.show()
```

```python
auc_score=roc_auc_score(y_test,model2.predict(x_test))
print(auc_score)
```

```python
# AUC-ROC for ---------Decision Tree----------
y_pred_prob1 = model1.predict_proba(x_test)[:,1]
fpr,tpr,thresholds=roc_curve(y_test,y_pred_prob1)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='Decision tree classifier')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Decision Tree Classifier')
plt.show()
auc_score=roc_auc_score(y_test,model1.predict(x_test))
auc_score
```

```python
# save the model(Random Forest Classifier)
import joblib
#save  the model as pickle in file
joblib.dump(model2,'Red_Wine_Classification.pkl')
```

By analyzing the above correlation matrix, it is seen that attribute "rings" have a good correlation with 'height, length and diameter'.

# Now analyzing the variation in correlation of "change in number of rings" with different attributes.
For this purpose using different values of the attribute "number of rings" 
1. less than 15
2. greater than 15
3. Less than 10 
4. Greater than 10

```python
# 1.  For Less than 15
df1 = df[df['Rings'] < 15]

plt.figure(figsize=(10, 8))
corr = df1.corr()
sns.heatmap(corr, annot=True)
```

```python
# 2.  For Greater than 15
df1 = df[df['Rings'] > 15]

plt.figure(figsize=(10, 8))
corr = df1.corr()
sns.heatmap(corr, annot=True)
```

```python
 # 3.  For Less than 10
df1 = df[df['Rings'] < 10]

plt.figure(figsize=(10, 8))
corr = df1.corr()
sns.heatmap(corr, annot=True)
```

```python
# 2.  For Greater than 10
df1 = df[df['Rings'] > 10]

plt.figure(figsize=(10, 8))
corr = df1.corr()
sns.heatmap(corr, annot=True)
```

1. From the above correlations maps we can generalize that the correlation of Attribute "rings" with other attributes decays to nearly "0" when we increase the number of rings from more than 10.
2. Hence we can say that abalones grows in size and weight until a certain age "nearly 10".(Age= Rings +1.5)
3. After this stage we can say that there is a least effect of age on these variables (Vice-Versa) or they depend more on environmental factors.  

```python
# Converting categorical data into numerical data by using label encoder

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['Sex'] = le.fit_transform(df['Sex'])
df.head()
```

```python
df.shape
```

```python
# Outlier detection and removal------by using Z-Score

from scipy import stats
df_outlier=df.copy()
z=np.abs(stats.zscore(df_outlier))
print(np.where(z>3))
```

This shows that there are outliers present in the dataset. which can also be seen from the description of the data done in the initial stage.

```python
# Removing the outliers
df_outlier = df_outlier[(z<3).all(axis=1)]
```

```python
df_outlier.shape
```

```python
# The above code removes around 150 rows as outliers from the dataset.
```

# Seperating feature variable and target variable

```python
x = df_outlier.drop('Rings',axis=1) # Input features
y = df_outlier.Rings # Target Variable
```

```python
# Splitting the data into Training and Validation Set

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=42)
```

```python
# Building a Model

from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
```

```python
# Training the model

rf.fit(x_train, y_train)
```

```python
# predicting on the validation data
pred = rf.predict(x_test)
```

```python
#Model evaluation
```

```python
from sklearn.metrics import mean_squared_error
print ('Root Mean Squared Error is:',np.sqrt(mean_squared_error(y_test, pred)))
```

```python
# Save the model
import joblib
# save the model as pickle in file
joblib.dump(rf,'Abalone_Model.pkl')
```
